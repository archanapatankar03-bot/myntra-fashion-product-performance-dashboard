```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
df=pd.read_csv("Myntra_Fashion.csv",low_memory=False)
df.head(10)
```

```python
#dimensions rows and columns
df.shape
```

```python
#column names
df.columns
```

```python
df.info()
```

```python
#check missing values discountprice,offer,ratings and reviews
df.isnull().sum()
```

```python
#check duplicate rows
df.duplicated().sum()
#if duplicates exists
df.drop_duplicates(inplace=True)
#verify
df.duplicated().sum()
#no duplicates
```

```python
#check data types
df.dtypes
```

```python
#phase 3
#long column names incovenient so make it short
df.rename(columns={
    "DiscountPrice (in Rs)": "DiscountPrice",
    "OriginalPrice (in Rs)": "OriginalPrice",
    "category_by_Gender": "Gender",
    "Individual_category": "SubCategory"
}, inplace=True)
```

```python
#verify whether names are shortened
df.columns
```

```python
#Clean text columns
#Remove unwanted spaces
df["BrandName"] = df["BrandName"].str.strip()
df["Category"] = df["Category"].str.strip()
df["SubCategory"] = df["SubCategory"].str.strip()
```

```python
#Standardize text
df["BrandName"] = df["BrandName"].str.title()
```

```python
#check missing ratings
df["Ratings"].isnull().sum()
```

```python
#fill with median values
median_rating = df["Ratings"].median()

df["Ratings"] = df["Ratings"].fillna(median_rating)
```

```python
#verify
df["Ratings"].isnull().sum()
```

```python
#fill missing reviews
df["Reviews"].isnull().sum()
```

```python
#fill null with 0
df["Reviews"] = df["Reviews"].fillna(0)
```

```python
#verify
df["Reviews"].isnull().sum()
```

```python
#convert review data type float to integer
df["Reviews"]=df["Reviews"].astype(int)
```

```python
##verify data type of review
df["Reviews"].dtypes
```

```python
#check discount price missing values
df["DiscountPrice"].isnull().sum()
```

```python
#fill with original price
df["DiscountPrice"] = df["DiscountPrice"].fillna(df["OriginalPrice"])
```

```python
#verify
df["DiscountPrice"].isnull().sum()
```

```python
#check missing values of discount offer
df["DiscountOffer"].isnull().sum()
```

```python
#fill missing values with no discount
df["DiscountOffer"] = df["DiscountOffer"].fillna("No Discount")
```

```python
#verify
df["DiscountOffer"].isnull().sum()
```

```python
#check for impossible values
df.describe()
```

```python
#Phase 4
#create new columns
#Discount Amount
df["DiscountAmount"] = df["OriginalPrice"] -df["DiscountPrice"]
```

```python
#Discount Percentage
df["DiscountPercentage"] = ((df["DiscountPrice"]/ df["OriginalPrice"]) * 100)
```

```python
#Revenue score
df["ReviewScore"] = (df["DiscountPrice"] * df["Reviews"])
```

```python
#Review new columns Discountamount,Discounpercentage and reviewscore
df.head()
```

```python
#Saved cleaned data
#This completes data cleaning
df.to_csv("Myntra_Cleaned.csv",index=False)
```
